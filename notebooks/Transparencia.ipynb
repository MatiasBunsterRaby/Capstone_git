```python
import pandas as pd
# Crear el DataFrame 'df_transparencia'
df_transparencia = pd.read_excel('./data/df_transparencia_.xlsx')
display(df_transparencia.head())
```

```python
df_transparencia.columns
```

```python
import matplotlib.pyplot as plt

# Convertir aos a formato numrico para facilitar operaciones
year_columns = [str(year) for year in range(2012, 2024)]

# Agrupar por 'cluster' y calcular los promedios de las columnas de aos
cluster_means = df_transparencia.groupby('cluster')[year_columns].mean()

# Crear un grfico de lneas
plt.figure(figsize=(10, 6))
for cluster in cluster_means.index:
    plt.plot(year_columns, cluster_means.loc[cluster], label=f'Cluster {cluster}')

# Configurar el grfico
plt.title('Promedios por Cluster (2012-2023)', fontsize=14)
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Promedio', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
