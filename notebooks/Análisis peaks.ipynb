```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Markdown, display
import glob
from PIL import Image
```

#### Cargar archivos

#### Cargamos los programas *.csv en 'df_programas_Chile'

```python
# Ruta a los archivos CSV
ruta_archivos = './data/Annotated_Chile/*.csv'

# Obtener la lista de archivos
archivos_csv = glob.glob(ruta_archivos)

# Verificar si se encontraron archivos
if not archivos_csv:
    print("No se encontraron archivos en la ruta especificada.")
else:
    print(f"Archivos encontrados: {archivos_csv}")

# Crear una lista para almacenar los DataFrames
dataframes = []

# Cargar los archivos al DataFrame
for archivo in archivos_csv:
    try:
        print(f"Leyendo archivo: {archivo}")
        df = pd.read_csv(
            archivo,
            encoding='utf-8',
            delimiter=',',
            skiprows=[1],  # Ignorar la primera fila después del encabezado
            names=['frase', 'frase_eng', 'cmp_code', 'eu_code'],  # Renombrar las columnas
            header=0,  # Ignorar los nombres originales de columnas
            keep_default_na=False,  # No interpretar valores como 'NA' como NaN
            na_values=[],  # No definir valores como NaN
            dtype={'cmp_code': str}  # Forzar la columna 'cmp_code' a tipo string
        )

        # Obtener el nombre del archivo sin ruta ni extensión
        nombre_archivo = archivo.split('/')[-1].replace('.csv', '')

        # Insertar la columna 'programa' como la primera columna
        df.insert(0, 'programa', nombre_archivo)

        # Agregar el DataFrame a la lista
        dataframes.append(df)
    except Exception as e:
        print(f"Error al leer {archivo}: {e}")

# Verificar si hay DataFrames para concatenar
if dataframes:
    df_programas_Chile = pd.concat(dataframes, ignore_index=True)
    print("DataFrame final:")
    print(df_programas_Chile.head())
else:
    print("No hay DataFrames para concatenar.")
```

#### Aseguremos que para cada programa em 'df_programas_Chile' existe en 'df_absolutos' un registro con el mismo 'party'_'date' 

```python
# Extract unique values of 'programa' from 'df_programas_Chile'
unique_programas = df_programas_Chile['programa'].unique()

# Initialize a list to store warnings
warnings = []

# Iterate over unique values of 'programa'
for programa in unique_programas:
    # Extract the leftmost 6 characters and rightmost 6 characters
    left_6 = programa[:6]
    right_6 = programa[-6:]
    
    # Validate if 'party' and 'date' exist in 'df_absolutos'
    if not ((df_absolutos['party'] == int(left_6)) & (df_absolutos['date'] == int(right_6))).any():
        warnings.append(f"Warning: No matching row found for 'party' == {left_6} and 'date' == {right_6}")

# Print warnings if any
for warning in warnings:
    print(warning)
```

#### En 'df_absolutos', cada columna 'per' que representa una variable que nos interesa, la multiplicamos por la columna 'total'

```python
# Create a list of column names that begin with 'per' and have a digit or an underscore in position 4
variables_per = [col for col in df_absolutos.columns if col.startswith('per') and (col[3].isdigit() or col[3] == '_')]

print(f"Total variables per: {len(variables_per)}")
```

```python
# Multiplicar cada columna en 'variables_per' por el valor en la columna 'total y dividir por 100'
for col in variables_per:
    df_absolutos[col] = (df_absolutos[col] * df_absolutos['total'] / 100).round(0).astype(int)
df_absolutos.head()
```

```python
# Define variables for the columns
tema = "DERECHOS HUMANOS"
col_1 = 'per201_1'
col_2 = 'per201_2'
# col_3 = 'per304'

# Select columns from df_absolutos_Chile_ponderados and df_porcentual_Chile_ponderados
df_abs_Chile_ddhh = df_absolutos_Chile_ponderados[['agno', col_1, col_2]].copy()
df_porcentual_Chile_ddhh = df_porcentual_Chile_ponderados[['agno', col_1, col_2]].copy()

# Add a new column 'Total' with the sum of columns in both datasets
df_abs_Chile_ddhh['Total'] = df_abs_Chile_ddhh[col_1] + df_abs_Chile_ddhh[col_2] #+ df_abs_Chile_orginarios[col_3]
df_porcentual_Chile_ddhh['Total'] = df_porcentual_Chile_ddhh[col_1] + df_porcentual_Chile_ddhh[col_2] #+ df_porcentual_Chile_orginarios[col_3]

# Create subplots (2 columns, 1 row) with shared x-axis
fig, (ax1, ax3) = plt.subplots(1, 2, figsize=(14, 6))

# ----------------- Primer gráfico: valores absolutos y porcentuales -----------------

# Plot the absolute values on the left axis
ax1.plot(df_abs_Chile_ddhh['agno'], df_abs_Chile_ddhh[col_1], label=f"{col_1} (Absoluto)", color='b')  # Azul
ax1.plot(df_abs_Chile_ddhh['agno'], df_abs_Chile_ddhh[col_2], label=f"{col_2} (Absoluto)", color='g')  # Verde
# ax1.plot(df_abs_Chile_orden['agno'], df_abs_Chile_orden[col_3], label=f"{col_3} (Absoluto)", color='r')  # Rojo
# Cambié el color de la línea de Total Absoluto a un color distinto para evitar colisión
ax1.plot(df_abs_Chile_ddhh['agno'], df_abs_Chile_ddhh['Total'], label=f"Total\n(Absoluto)", color='black')  # Negro

ax1.set_xlabel('Año')
ax1.set_ylabel('Valores Absolutos')
ax1.set_title(f'PAÍS: CHILE - Valores Absolutos y Porcentuales\n{tema}')
ax1.grid(True)
ax1.legend(loc="upper left")

# Create a second y-axis for the porcentual values
ax2 = ax1.twinx()
ax2.plot(df_porcentual_Chile_ddhh['agno'], df_porcentual_Chile_ddhh[col_1], label=f"{col_1} (Porcentual)", linestyle='--', color='b')  # Azul
ax2.plot(df_porcentual_Chile_ddhh['agno'], df_porcentual_Chile_ddhh[col_2], label=f"{col_2} (Porcentual)", linestyle='--', color='g')  # Verde
#ax2.plot(df_porcentual_Chile_orden['agno'], df_porcentual_Chile_orden[col_3], label=f"{col_3} (Porcentual)", linestyle='--', color='r')  # Rojo
# Cambié el color de la línea de Total Porcentual a un color diferente
ax2.plot(df_porcentual_Chile_ddhh['agno'], df_porcentual_Chile_ddhh['Total'], label=f"Total\n(Porcentual)", linestyle='--', color='purple')  # Púrpura

ax2.set_ylabel('Valores Porcentuales')
ax2.legend(loc="upper right")

# ----------------- Segundo gráfico: sólo los Totales -----------------

# Plot the "Total" lines for both datasets in the second subplot
ax3.plot(df_abs_Chile_ddhh['agno'], df_abs_Chile_ddhh['Total'], label=f"Total\n(Absoluto)", color='orange')  # Naranja para total absoluto
ax3.set_xlabel('Año')
ax3.set_ylabel('Valores Absolutos')
ax3.grid(True)

# Create a second y-axis for the porcentual "Total" line
ax4 = ax3.twinx()
ax4.plot(df_porcentual_Chile_ddhh['agno'], df_porcentual_Chile_ddhh['Total'], label=f"Total\n(Porcentual)", color='brown')  # Marrón para total porcentual
ax4.set_ylabel('Valores Porcentuales')

# Set titles and legends for the second subplot
ax3.set_title(f'PAÍS: CHILE - Totales Absolutos y Porcentuales\n{tema}')
ax3.legend(loc="upper left")
ax4.legend(loc="upper right")

texto_md = """<div style="display: flex; justify-content: space-between;">
  <img src="../assets/per201_1.png" alt="" width="550"></div>
  <img src="../assets/per201_2.png" alt="" width="650">
</div>
"""
display(Markdown(texto_md))

# Show the plot
plt.tight_layout()  # Adjust layout to fit both plots
plt.show()
#________________________________________________________
# GRAFICAMOS LOS ABSOLUTOS SIN PONDERAR
# Crear la lista 'agnos_a_analizar'
agnos_a_analizar = [1989, 1993, 1999, 2005, 2009, 2013, 2017, 2021]

# Crear el dataframe 'df_abs_1989_2021' como copia de 'df_absolutos'
df_abs_1989_2021 = df_absolutos.copy()

# Crear una lista para almacenar los resultados
resultados = []

# Iterar sobre los valores en 'agnos_a_analizar'
for agno in agnos_a_analizar:
    # Filtrar el dataframe por el año actual
    df_filtered_year = df_abs_1989_2021[df_abs_1989_2021['agno'] == agno]
    
    # Iterar sobre cada partido en el dataframe filtrado
    for index, row in df_filtered_year.iterrows():
        party = row['party']
        candidatename = row['candidatename']
        per201_1_value = row['per201_1']
        per201_2_value = row['per201_2']
        
        # Agregar los valores a la lista de resultados
        resultados.append({
            'agno': agno,
            'party': party,
            'candidatename': candidatename,
            'per201_1': per201_1_value,
            'per201_2': per201_2_value
        })

# Crear el dataframe con los resultados
df_resultados = pd.DataFrame(resultados)

# Crear una nueva columna 'Total' que sea la suma de 'per201_1' y 'per201_2'
df_resultados['Total'] = df_resultados['per201_1'] + df_resultados['per201_2']

# Agrupar por 'agno' y sumar los valores de 'Total' para cada año
df_agg = df_resultados.groupby('agno')['Total'].sum().reset_index()

# Crear la figura con dos subplots: uno para el gráfico y otro para la imagen
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios': [2, 2]})

# Gráfico de línea en el subplot izquierdo
ax1.plot(df_agg['agno'], df_agg['Total'], marker='o', linestyle='-', color='b')
ax1.set_xlabel('Año')
ax1.set_ylabel('Menciones')
ax1.set_title('DERECHOS HUMANOS\n (201_1 + 201_2 absolutos sin ponderar)')
ax1.grid(True)

# Cargar y redimensionar la imagen utilizando PIL
img = Image.open('../assets/peak_ddhh.png')
img_resized = img.resize((800, 800))  # Cambia las dimensiones a 500x500 píxeles, ajusta según necesites

# Convertir la imagen redimensionada a un formato compatible con matplotlib
img_resized_np = np.array(img_resized)

# Mostrar la imagen redimensionada en el subplot derecho
ax2.imshow(img_resized_np)
ax2.axis('off')  # Quitar los ejes de la imagen

# Ajustar el diseño de la figura para que los subplots no se solapen
plt.tight_layout()
plt.show()

# ______________________________________________________
#
# TABULAMOS
from tabulate import tabulate
df_grouped = df_resultados.groupby(['agno', 'candidatename']).agg({'per201_1': 'sum', 'per201_2': 'sum'}).reset_index()
df_grouped['Total'] = df_grouped['per201_1'] + df_grouped['per201_2']
table = df_grouped.values.tolist()
headers = ['Año', 'Candidato', 'per201_1', 'per201_2', 'Total']
print(tabulate(table, headers, tablefmt='pretty'))

# Guardamos el dataframe en un archivo de Excel llamado 'ddhh_peak.xlsx'
df_grouped.to_excel('ddhh_peak.xlsx', index=False)
```
